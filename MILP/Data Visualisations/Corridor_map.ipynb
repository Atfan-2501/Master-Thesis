```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# 1. Load convergence data
# -----------------------------
file_path = "../Results/Experiment 1/model_output/benders_convergence.xlsx"  # adjust path if needed

df = pd.read_excel(file_path)

# Make sure iterations are sorted
df = df.sort_values("iteration")

# If you only want iterations from the beginning of the log:
# df = df[df["iteration"] >= 1]

# -----------------------------
# 2. Plot relative gap
# -----------------------------
sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.lineplot(
    data=df,
    x="iteration",
    y="relative_gap",
    marker="o"
)

plt.xlabel("Benders iteration")
plt.ylabel("Relative optimality gap")

# Optionally show gap in percentage on y-axis:
# import matplotlib.ticker as mtick
# plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

plt.tight_layout()

# -----------------------------
# 3. Save & show
# -----------------------------
os.makedirs("figures", exist_ok=True)
out_path = os.path.join("figures", "benders_gap_evolution.png")
plt.savefig(out_path, dpi=300)
print(f"✅ Figure saved to: {out_path}")

plt.show()
```
